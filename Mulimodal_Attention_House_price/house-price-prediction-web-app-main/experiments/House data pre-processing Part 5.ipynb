## Fill in all the missing values with 0 for numerical values, and NotListed for categorical values

```python
df.maintenanceFees=df.maintenanceFees.fillna(0)
df.landSize=df.landSize.fillna(0) #if landsize is not listed, fill in value with 0 
```

```python
df.maintenanceManagementCompany=df.maintenanceManagementCompany.fillna("notlisted")
df.maintenanceManagementCompany=df.maintenanceManagementCompany.str.replace("notlisted", "NotListed")
```

```python
df.basementType=df.basementType.fillna("None")
df.basementType=df.basementType.str.replace("None", "NotListed")

df.basementFinish=df.basementFinish.fillna("None")
df.basementFinish=df.basementFinish.replace("None", "NotListed")

df.heatingType1=df.heatingType1.fillna("NotListed")
df.heatingType2=df.heatingType2.fillna("NotListed")
df.heatingEnergy1=df.heatingEnergy1.fillna("NotListed")
df.heatingEnergy2=df.heatingEnergy2.fillna("NotListed")
```

```python
df.featureLotSlope=df.featureLotSlope.fillna("NotListed")
df.featureDriveway=df.featureDriveway.fillna("NotListed")
df.featureLotPositionType=df.featureLotPositionType.fillna("NotListed")
df.featureOutdoorAreaType=df.featureOutdoorAreaType.fillna("NotListed")
df.featureOutdoorLandscape=df.featureOutdoorLandscape.fillna("NotListed")
df.featureAdditionalFacility=df.featureAdditionalFacility.fillna("NotListed")
```

```python
df.cooling=df.cooling.str.replace("None", "NotListed")
```

```python
df.to_csv(output_folder+"/"+"all_after_preprocessing5.csv", index=False)
```
